```python
def calculate_dot_product(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Vectors must have the same length")
    
    return sum([v1[i] * v2[i] for i in range(len(v1))])
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([3, 4, 5])

manual_dot_product = calculate_dot_product(v1, v2)

np_dot_product = np.dot(v1, v2)

assert manual_dot_product == np_dot_product
```

```python
def calculate_hadamard_product(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Vectors must have the same length")
    
    return [v1[i] * v2[i] for i in range(len(v1))]
```

```python
manual_hadamard_product = calculate_hadamard_product(v1, v2)
np_hadamard_product = v1 * v2

assert [i[0] == i[1] for i in enumerate(zip(manual_hadamard_product, np_hadamard_product))]
```

```python
def calculate_outer_product(u, v):    
    result = []
    
    for i in u:
        row = []
        for j in v:
            row.append(i * j)
        result.append(row)
    
    return np.array(result)
    
    # return np.array([[v1[i] * v2[j] for j in range(len(v2))] for i in range(len(v1))])
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([3, 4, 5, 2])

manual_outer_product = calculate_outer_product(v1, v2)
np_outer_product = np.outer(v1, v2)

assert np.all(manual_outer_product == np_outer_product)
```

```python
def determine_vector_dependence(v: np.array, u: np.array) -> bool:
    # Create a matrix from the vectors
    matrix  = np.vstack([v, u]).T
    
    # Calculate the rank of the matrix
    matrix_rank = np.linalg.matrix_rank(matrix)
    
    # If the rank is less than 2, the vectors are dependent
    print(matrix_rank)
    return matrix_rank < 2


# Independent vectors
v1 = np.array([1, 2, 3])
v2 = np.array([3, 4, 5])

# Dependent vectors
v3 = np.array([1, 2, 3])
v4 = np.array([3, 6, 9])

# Test the function
assert not determine_vector_dependence(v1, v2)
assert determine_vector_dependence(v3, v4)
```

## Matrix Multiplication

Matrix multiplication animation: https://matrixmultiplication.xyz/

The most important rule of matrix multiplication: the numbers of columns in the left matrix must match the number of rows in the right matrix. If the condition is met the multiplication can be done and the result will be the number of rows of the left matrix vs the number of column of the right matrix.

3 X 2 @ 2 X 4 --- ?

3 X 4 @ 3 X 4 --- ?

3 X 5 @ 5 X 4 --- ?

7 X 11 @ 11 X 4 --- ?

(3 X 4 ).transpose() @ 4 X 3 --- ?

```python
def gen_matrix(sizes: tuple) -> np.array:
    
    matrix = np.arange(1, np.prod(sizes) +1).reshape(sizes)
    
    return matrix
```

```python
def multiplie_matrices(matrix_a_size: tuple, matrix_b_size: tuple):
    
    try:
    
        m_a = gen_matrix(matrix_a_size)
        m_b = gen_matrix(matrix_b_size)    

        return m_a @ m_b
    
    except ValueError as e:
        print('something went wrong')
        print(e)
```

```python
A = np.array([[4, 1],
              [2, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
v = np.array([0.70710678, 0.70710678])

lambda_val = 5

Av = A @ v

lambda_v = lambda_val * v

print("A @ v:\n", Av)
print("lambda * v:\n", lambda_v)

print("Are A @ v and lambda * v approximately equal?", np.allclose(Av, lambda_v))
```
