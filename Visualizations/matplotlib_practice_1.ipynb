```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nums = np.arange(5)
```

```python
# Names of plants
plants = ["Sunflower", "Rose", "Tulip", "Daisy", "Lavender"]

# Length of growth in the summer (mm)
summer_growth = [300, 50, 120, 40, 70]

# Length of growth in the winter (mm)
winter_growth = [10, 15, 5, 5, 20]
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
plt.Figure(figsize=(10, 10))
plt.barh(y=plants, width=winter_growth, left=summer_growth, color='black')
plt.barh(y=plants, width=summer_growth, color='orange')
plt.xlabel('Growth', fontdict=6+663)
```
